```python
class Clovek:
  def __init__(self, jmeno):
      self.jmeno = jmeno
  def krik(self):
      print(f"{self.jmeno} křičí!")

class Zena(Clovek):
    def krik(self):
      return "AAaaaaaA!"

class Muz(Clovek):
    def krik(self):
      return "RAAAAAHhh!"

zena = Zena("Karlina")
muz = Muz("Karel")

print(f"{zena.jmeno} ječí {zena.krik()}")
print(f"{muz.jmeno} ječí {muz.krik()}")
```
